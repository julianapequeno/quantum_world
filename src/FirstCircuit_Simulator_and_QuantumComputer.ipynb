```python
import numpy as np
from qiskit import *
```

```python
qc = QuantumCircuit(2,2) #quantum and classical registers with 2 qubits each
```

```python
qc.measure([0,1],[0,1]) # mapping quantum register to classical register
```

```python
qc.draw('mpl')
```

```python
# Import Aer for simulating a quantum circuit in a classical computer
from qiskit import Aer
#getting classical quantum simulator
simulator = Aer.get_backend('qasm_simulator')
```

```python
job = execute(qc,simulator,shots=1000) #shots = how many times do you want to repeat the experiment
```

```python
result = job.result()
counts = result.get_counts(qc)
print('\nCount:',counts)
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

```python
#executing in a real quantum computer
from qiskit_ibm_provider import IBMProvider

# Save account credentials.
# IBMProvider.save_account(token=MY_API_TOKEN)

# Load a previously saved account.
provider = IBMProvider()

# Select a different hub/group/project.
provider = IBMProvider(instance="ibm-q/open/main")
backends = provider.backends(simulator=False, operational=True)
backend = backends[0]
```

```python
job = execute(qc, backend, shots=1000)
```

```python
result = job.result()
counts = result.get_counts(qc)
print('\nCount:',counts)
```

```python
plot_histogram(counts)
```
